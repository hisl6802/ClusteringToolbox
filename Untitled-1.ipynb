```python
import numpy as np 
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering as AC
import matplotlib.pyplot as plt



data,labels = make_blobs(1000, 2,cluster_std=5,random_state=42)
out = AC(3).fit(data)
#getting the plot object 
fig,ax = plt.subplots(1,2,figsize=(10,5))
ax[0].scatter(data[:,0],data[:,1])#,c=labels)
ax[0].spines[['top','right']].set_visible(False)
yticks = [-20,-10,0,10,20,30]
ax[0].set_yticks(yticks, labels=yticks,font='Arial',fontsize=12)
xticks = [-20,-15,-10,-5,0,5,10,15]
ax[0].set_xticks(xticks,labels=xticks,font="Arial",fontsize=12)

ax[1].scatter(data[:,0],data[:,1],c=out.labels_)#,c=labels)
ax[1].spines[['top','right']].set_visible(False)
yticks = [-20,-10,0,10,20,30]
ax[1].set_yticks(yticks, labels=yticks,font='Arial',fontsize=12)
xticks = [-20,-15,-10,-5,0,5,10,15]
ax[1].set_xticks(xticks,labels=xticks,font="Arial",fontsize=12)

plt.savefig('ClusteringBadResults.png',dpi=600)
```
