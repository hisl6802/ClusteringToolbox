```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.spatial.distance import pdist

import sklearn.datasets as ds

#get the data from the ECCO_Check
data = pd.read_excel("ECCO_check.xlsx")
data=data.drop(0)
data=data.drop('m/z',axis=1)
data=data.drop('rtmed',axis=1)
```

```python
data_np =data.to_numpy(dtype=float)
print(data_np)
np.log10(data_np+0.1)

test = ds.make_blobs(1000,2,centers=4)

test = test[0]
plt.scatter(test[:,0],test[:,1])
```

```python
dists = pdist(test)

plt.hist(dists,bins=100);
```
